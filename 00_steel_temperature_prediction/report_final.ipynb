## Отчет

### Введение

#### Задача проекта

Чтобы оптимизировать производственные расходы, металлургический комбинат ООО «Так закаляем сталь» решил уменьшить потребление электроэнергии на этапе обработки стали. Нужно построить модель, которая предскажет температуру стали.

#### Описание этапа обработки
Сталь обрабатывают в металлическом ковше вместимостью около 100 тонн. Чтобы ковш выдерживал высокие температуры, изнутри его облицовывают огнеупорным кирпичом. Расплавленную сталь заливают в ковш и подогревают до нужной температуры графитовыми электродами. Они установлены в крышке ковша. 

Из сплава выводится сера (десульфурация), добавлением примесей корректируется химический состав и отбираются пробы. Сталь легируют — изменяют её состав — подавая куски сплава из бункера для сыпучих материалов или проволоку через специальный трайб-аппарат (англ. tribe, «масса»).

Перед тем как первый раз ввести легирующие добавки, измеряют температуру стали и производят её химический анализ. Потом температуру на несколько минут повышают, добавляют легирующие материалы и продувают сплав инертным газом. Затем его перемешивают и снова проводят измерения. Такой цикл повторяется до достижения целевого химического состава и оптимальной температуры плавки.

Тогда расплавленная сталь отправляется на доводку металла или поступает в машину непрерывной разливки. Оттуда готовый продукт выходит в виде заготовок-слябов (англ. slab, «плита»).

#### Описание данных
Данные состоят из файлов, полученных из разных источников:
- `data_arc.csv` — данные об электродах;
- `data_bulk.csv` — данные о подаче сыпучих материалов (объём);
- `data_bulk_time.csv` — данные о подаче сыпучих материалов (время);
- `data_gas.csv` — данные о продувке сплава газом;
- `data_temp.csv` — результаты измерения температуры;
- `data_wire.csv` — данные о проволочных материалах (объём);
- `data_wire_time.csv` — данные о проволочных материалах (время).

Во всех файлах столбец `key` содержит номер партии. В файлах может быть несколько строк с одинаковым значением `key`: они соответствуют разным итерациям обработки.

**Целевое значение температуры** это последний замер температуры в партии.
Еще по условию задачи для модели будет доступно первое измерение.

**Требовалось чтоб целевая метрика MAE была не более 8.7**

### Исследовательский анализ данных

После загрузки данных был проведен аналих данных по каждой полученной таблице. Для удобства изучения таблиц была созданна функция *get_info*. С помощью этой функции получали следующую информацию по каждой таблице:
- Заглавие и первые пять сторк таблицы
- Информацию по каждому столбцу
- Статистические параметры по каждому столбцу
- Проводилась проверка на дубликаты
- Считалось количество уникальных партий по ключу *key*
- Расчитывалась корреляционная матрица

Далее приведем основные моменты по каждой таблице

#### Данные об электродах
В таблице приводятся время старта и окончания нагрева партии сплава конкретным электродом, активная и реактивная мощность электрода. Всего 14876 строк, 3214 уникальных партий. 

Из отклонений отмеченно единственное некорректное значения реактивной мощности -715 от которого избавились. Как и полагается, наблюдается сильная корреляция между активной и реактивной мощностями. В дальнейшем, для обучения модели использовалось значение энергии нагрева сплава, так как в ней уже есть сведенья о времени нагрева и активной мощьности. 

#### Результаты измерения температуры
В таблице приводятся время измерения температуры партии и соответствующая температура. 
Строк 15907, но есть пропуски в данных о температуре.  Дубликаты отсутствуют.
Количество уникальных номеров партий было 3216. После очистки данных от сомнительных партий (слишком низкая температура, отсутствие данных) осталось 2470. При сведении полученных таблиц в таблицу признаков это было учтенно.
Как итог, полученна таблица в 2470 строки с первичной температурой и целевым признаком для каждой партии.
В эту таблицу не вошли:
- Партии с температурой ниже температуры плавления стали.
- Партии с отсутствием первого или последнего замера, то есть без целевого признака.

#### Данные о подаче сыпучих материалов
Данные о подаче сыпучих материалов были предсталенны в двух таблицах: в виде объёма каждого типа материала и времени подачи. В модели участвовали данные по объёму сыпучих материалов. Количество строк соответсвовало количеству учтенных партий в этой таблице 3129. Было 15 сыпучих материалов: от bulk_1 до	bulk_15.
Пропуски в данной таблице означают то, что данные сыпучий материал не подавался в партию сплава. Поэтому пропуски были заполнены нулями. Дубликатов в данных нет.
Некоторые материали подавались очень редко, например bulk_8	только в одну партию из 3129.
Ещё было отмеченно, что некоторые дабавки добавляются совместно с другими и объём их коррелирует. Например, добавка bulk_1 всегда добавляется с bulk_13, а добавка bulk_7 всегда добавляется с bulk_14.

#### Данные о продувке сплава газом
Это простая таблица, состоящая из двух столбцов: номер партии и объём газа. Количество учтеных партий 3239
Объём газа для продувания сплава лежит в диапазоне от 0.008 до 78 с медианным значением 9.8 и средним 11.

#### Данные о проволочных материалах

Данные о проволочных материалах, как и данные о подаче сыпучих материалов, в двух таблицах: в виде объёма каждого типа материала и времени подачи.
Количество учтеных партий 3081, что меньше чем для сыпучих материалов 3129 и газа 3239. 
Некоторые добавки очень редко использовались. Например, добавка wire_5 добавлялась только в одну партию, а добавки wire_4, wire_7 и wire_8 добавлялись в менее чем 20 партий. 



### Подготовка данных

- Расчитали энегию нагрева каждого сплава в виде суммы энергий по его активной мощности электрода и времени его включения. 
- Формирование таблицы признаков осуществлялось путем добавления данных и полученных таблиц. Партии, для которых не было данных в любой из таблиц исключались. В итоге осталось количество уникальных партий 2324, что чуть меньше чем в очищенной таблице результатов измерения температуры (2470).
- К сожалению, единственная партия, когда добавлялся *wire_5*, не вошла в итоговую талицу, и поэтому признак *wire_5* стал бесполезен для обучения - все данные нулевые. Но чтоб не потерять этот признак в дальнейшем, объединили его с другими редко встречающимися признаками: *wire_4*, *wire_7*, *wire_8*, *wire_9*.
- В сыпучих материалах тоже есть редко используемые добавки: *bulk_2*, *bulk_7*, *bulk_8*, *bulk_9*, *bulk_13*. Эти признаки также объеденили в один bulk_278913, чтоб уменьшить количество в основном нулевых признаков.

Далее проанализаровали полученную таблицу признаков. И из чего можно отметить следующее:
- В полученной таблице признаков можно выделить положительную корреляцию между затраченной энергией и добавлением добавок *bulk_12* (0.48) и *bulk_14* (0.44), и отрицательную с первичной температурой *first_temp* (-0.42). 
- Целевой признак *last_temp* очень сильных корреляций ни с чем не имеет.

В итоге у нас получилось 19 признаков плюс целевой.

Полученную таблицу признаков разбили на две выборки:
- Тренировочную, размером 1859 на 19,
- и тестовую, размером 465 на 19.

### Построение модели машинного обучения

1. Построили несколько моделей машинного обучения:
    - Случайный лес
    - LightGBMRegressor
    - CatBoostRegressor
    - XGBRegressor
    
    Проверку работы модели на этапе тренировки модели и подбора гиперпараметров осуществляли способом перекрёстной валидации используя *GridSearchCV*.

#### Выбор лучшей модели 
Выбрали модель, показывающую наилучшее качество.

| **MAE на тренировочной выборке** | |
| :------------ | :-------- |
| CatBoost | 5.798615 |
| XGBoost | 5.894100 |
| Случайный_лес	| 5.904245 |
| Light_GBM | 5.971736 |

У всех четырех лучших моделей МАЕ меньше 6. Разница между самой лучшей моделью и худшей всего 0.17.
Лучшей моделью с результатом МАЕ=5.798615 на тренировочной выборке стала модель на основе градиентного бустинга CatBoost, оторвавшись от остальных на одну десятую.

#### Оценка лучшей модели на тестовой выборке
Лучшая модель CatBoostRegressor на тестовой выборке достигла результат **МАЕ = 6.19**.

Параметры лучшей модели CatBoostRegressor следующие:
- 'depth': 6, 
- 'iterations': 300, 
- 'l2_leaf_reg': 2, 
- 'learning_rate': 0.05

Хочется отметить, что для трех лидирующих моделей начальная температура сплава *first_temp* и затраченная на нагрев энергия *active_energy* являются самыми важными признаками. А признаки *wire_1* и *bulk_6* входят во все эти пятерки.

3. Признаки отражающие количество добавленных материалов и объёма прокаченного газа оказались менее важны.

### Общий вывод

Была поставлена задача создать модель машинного обучения, прогнозирующую температуру сплава перед подачей его на розлив. В нашем распоряжении были таблицы содержащие данные о замерах температуры, о нагреве сплава, о добавках в сплав и о продувке сплава газом. Эти данные были изучены, очищены, из них были полученны необходимые признаки для моделирования.

На тренировачном наборе данных, были обучены несколько моделей с подбором гиперпараметров. Лучшее качество показала модель `CatBoostRegressor`. При тестировании данная модель также показала хорошее качество MAE=6.19.

Анализ важности признаков показал, что для прогноза наиболее важны начальная температура сплава и энергия нагрева. Следовательно, регулируя начальную температуру, можно уменьшить расходы на нагрев. Вероятно также можно повлиять на расход электроэнергии изменением температуры подаваемых материалов и газа.
